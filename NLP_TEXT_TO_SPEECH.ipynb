```python
# first lets install libraries we have already install gtts with pip so m not reinstalling it
from gtts import gTTS
import IPython.display as ipd
from io import BytesIO
```

```python
# lets introduce txt........ in my case m adding my self introduction

introduction = """
I am Sonkar Abhishek from Mumbai, and I'd like to introduce myself. I graduated from SSJCOE,
which is affiliated with Mumbai University, in June 2021. I've been actively employed since
October 2022, gaining valuable work experience. Initially, my journey began at Digifabb,
a 3D printing manufacturing company, and currently, I hold the position of a production engineer
at Folds Design Studio, an architectural firm. In my free time, I am passionate about watching
anime, as well as engaging in outdoor sports, particularly football and cricket. Additionally,
I have been involved in cryptocurrency trading, where I have successfully applied my analytical
skills to achieve profitable results. I have harnessed these skills effectively in the world
of trading, generating substantial profits. Moreover, I possess a solid understanding of tools
such as Tableau and Google Analytics, as well as data visualization and machine learning concepts.
Thank you for considering my application. I look forward to the possibility of joining the Star
Student Program and contributing to its success.
"""

```

```python
# i am Splitting  the text into chunks
chunk_size = 400
chunks = [introduction[i:i+chunk_size] for i in range(0, len(introduction), chunk_size)]

```

```python
# lets generate audio for each chunks
full_audio = None
for chunk in chunks:
    tts_chunk = gTTS(text=chunk, lang='en', slow=False)
    if full_audio is None:
        full_audio = BytesIO()
        tts_chunk.write_to_fp(full_audio)
    else:
        tts_chunk.write_to_fp(full_audio)

```

```python
# Save the full audio to a file
full_audio.seek(0)
with open('full_intro.mp3', 'wb') as f:
    f.write(full_audio.read())

```

```python
# lets Play the full audio
ipd.display(ipd.Audio('full_intro.mp3', autoplay=True))
```
